# Split Data into Train and Test

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into training and testing sets
X_train, X_test, y_train_min, y_test_min, y_train_max, y_test_max = train_test_split(df.iloc[:, :-2], df.iloc[:, -2], df.iloc[:, -1], test_size=0.3, random_state=42)
```

# Creating and Training the linearRegression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# Creating a linear regression model

```python
model = LinearRegression()
```

# Predicting the minimum and maximum temperature for the next day

```python
model.fit(X_train, y_train_min)
y_pred_min = model.predict(X_test)
# Print the predicted temperature
print("Predicted temperature: ", y_pred_min)
```

```python
model.fit(X_train, y_train_max)
y_pred_max = model.predict(X_test)
# Print the predicted temperature
print("Predicted temperature: ", y_pred_max)
```

# Calculating the mean squared error

```python
mse_min = mean_squared_error(y_test_min, y_pred_min)
mse_max = mean_squared_error(y_test_max, y_pred_max)
```

```python
print(f'The mean squared error for the predicted minimum temperature is {mse_min:.2f}')
print(f'The mean squared error for the predicted maximum temperature is {mse_max:.2f}')
```
